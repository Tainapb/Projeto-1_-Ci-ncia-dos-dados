```python
import re 


def cleanup(text):
    """
        Função de limpeza muito simples que troca alguns sinais básicos por espaços
    """
    #import string
    punctuation = '[",!-.:?;çã'']' # Note que os sinais [] são delimitadores de um conjunto.
#     punctuation = '[!-.:?;]'
    pattern = re.compile(punctuation)
    
    text_subbed = re.sub(pattern, '', text)
    return text_subbed
```

```python
def limpa_prep(texto): 
    lista=['a','e','é','i','o','u','ou', 'ante', 'após', 'até', 'se','com', 'conta', 'de', 'desde', 'em', 'entre', 'para', 'per', 'perante', 'por', 
          'sem', 'sob', 'sobre', 'tras', 'na', 'no', 'as','os', 'que','dos','das', 'um', 'uma', 'umas', 'da', 'do', 'o', 'as', 'pra','nas', 'nos', 'ao', 'seu','sua', 'suas', 'seus', 'às']
    lista_frase=texto.split()
    result=[palavra for palavra in lista_frase if palavra.lower() not in lista ]
    retorno=' '.join(result)
    return retorno
```
